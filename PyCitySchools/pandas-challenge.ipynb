```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
num_stu = len(student_data)
pass_math = (sum(student_data["math_score"] >=70) / num_stu) * 100
pass_read = (sum(student_data["reading_score"] >=70) / num_stu) * 100
stu_pass = ((len(student_data.loc[(student_data['math_score']>=70) & (student_data['reading_score']>= 70)])) / num_stu) * 100

dist_sumry = [
    {"Total Schools":len(school_data), 
     "Total Students":num_stu, 
     "Total Budget":sum(school_data["budget"]), 
     "Average Math Score":student_data["math_score"].mean(),
     "Average Reading Score":student_data["reading_score"].mean(),
     "% Passing Math":pass_math, 
     "% Passing Reading":pass_read, 
     "% Overall Passing":stu_pass}]
dist_sumry_df = pd.DataFrame(dist_sumry)
pd.options.display.float_format = "{:,.6f}".format
dist_sumry_df
```
